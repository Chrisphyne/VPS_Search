````python
import pandas as pd
from langchain_core.prompts import ChatPromptTemplate
from langchain_experimental.tools import PythonAstREPLTool
from langchain_core.output_parsers.openai_tools import JsonOutputKeyToolsParser

# Assume df is ksb_df_1
# df = ksb_df_1
# df_1 = df[df['year'] == 2025][['Crop Yield (tonnes/ha)', 'sucrose content']]
# df_2 = df[df['year'] == 2025][['sucrose content', 'Production Quantity (tonnes)']]

df_1 = ksb_df_1
df_2 = ksb_df_2

tool = PythonAstREPLTool(locals={"df_1": df_1, "df_2": df_2})
llm_with_tool = llm.bind_tools(tools=[tool], tool_choice=tool.name)
parser = JsonOutputKeyToolsParser(key_name=tool.name, first_tool_only=True)

df_template = """```python
{df_name}.head().to_markdown()
>>> {df_head}
```"""
df_context = "\n\n".join(
    df_template.format(df_head=_df.head().to_markdown(), df_name=df_name)
    for _df, df_name in [(df_1, "df_1"), (df_2, "df_2")]
)

system = f"""You have access to a number of pandas dataframes. 
Here is a sample of rows from each dataframe and the python code that was used to generate the sample:

{df_context}

Given a user question about the dataframes, write the Python code to answer it. 
Don't assume you have access to any libraries other than built-in Python ones and pandas. 
Make sure to refer only to the variables mentioned above."""
prompt = ChatPromptTemplate.from_messages([("system", system), ("human", "{question}")])

chain = prompt | llm_with_tool | parser | tool
result = chain.invoke(
    {
        "question": " Which factory performed the best in 2025 in terms of yeilds and sucrose content?"
    }
)
print(result)
````

````python
import pandas as pd
import json

# Initialize results dictionary for JSON
results = {}

# Initialize Markdown output
markdown_output = "# Comprehensive Insights for Kenyan Sugar Board (2024-2025)\n\n"

# Helper function to convert NumPy types to Python native types
def convert_to_native(obj):
    if isinstance(obj, (pd.Int64Dtype, pd.Float64Dtype, pd.core.arrays.integer.Int64Dtype)):
        return int(obj)
    elif isinstance(obj, (pd.Float64Dtype, pd.core.arrays.floating.Float64Dtype)):
        return float(obj)
    elif isinstance(obj, pd.Series):
        return obj.apply(convert_to_native).to_dict()
    elif isinstance(obj, pd.DataFrame):
        return obj.applymap(convert_to_native).to_dict(orient='records')
    elif isinstance(obj, dict):
        return {k: convert_to_native(v) for k, v in obj.items()}
    elif isinstance(obj, list):
        return [convert_to_native(item) for item in obj]
    return obj

# Insight 1: Lowest crop yield region in 2025
df_2025 = df[df['year'] == 2025]
avg_yield_by_region = df_2025.groupby('region')['Crop Yield (tonnes/ha)'].mean().sort_values()
lowest_yield_region = avg_yield_by_region.index[0]
lowest_yield_value = float(avg_yield_by_region.iloc[0])  # Convert to float
details_1 = df_2025[df_2025['region'] == lowest_yield_region][['factory', 'variety', 'Crop Area (hectares)', 'Cane Deliveries (tonnes)', 'sucrose content', 'Crop Yield (tonnes/ha)']]
avg_sucrose_1 = float(details_1['sucrose content'].mean(skipna=True))
avg_crop_area_1 = float(details_1['Crop Area (hectares)'].mean(skipna=True))
variety_counts_1 = details_1['variety'].value_counts().to_dict()
factories_1 = details_1['factory'].unique().tolist()
min_yield_1 = float(details_1['Crop Yield (tonnes/ha)'].min())
max_yield_1 = float(details_1['Crop Yield (tonnes/ha)'].max())
insights_1 = []
if lowest_yield_value < 50:
    insights_1.append("Low yield may indicate poor soil quality or inadequate irrigation.")
if avg_sucrose_1 < 12:
    insights_1.append("Low sucrose content suggests lower-quality cane.")
if avg_crop_area_1 < 300:
    insights_1.append("Small crop areas may limit yield efficiency.")
if len(variety_counts_1) > 1:
    insights_1.append("Multiple varieties used; evaluate their performance.")
results['lowest_crop_yield_2025'] = {
    'region': lowest_yield_region,
    'average_yield': round(lowest_yield_value, 2),
    'insights': insights_1,
    'details': details_1.to_dict(orient='records')
}
markdown_output += f"""
## 1. Lowest Crop Yield Region in 2025
- **Region**: {lowest_yield_region}
- **Average Yield**: {round(lowest_yield_value, 2)} tonnes/ha
- **Insights**:
{''.join([f'  - {i}\n' for i in insights_1])}
- **Details**:
{details_1.to_markdown(index=False)}
\n"""

# Insight 2: Highest production factory in 2024
df_2024 = df[df['year'] == 2024]
total_prod_by_factory = df_2024.groupby('factory')['Production Quantity (tonnes)'].sum().sort_values(ascending=False)
top_factory = total_prod_by_factory.index[0]
top_prod_value = float(total_prod_by_factory.iloc[0])  # Convert to float
details_2 = df_2024[df_2024['factory'] == top_factory][['region', 'variety', 'Crop Yield (tonnes/ha)', 'sucrose content']]
avg_yield_2 = float(details_2['Crop Yield (tonnes/ha)'].mean(skipna=True))
avg_sucrose_2 = float(details_2['sucrose content'].mean(skipna=True))
insights_2 = []
if avg_yield_2 > 80:
    insights_2.append("High crop yield likely contributes to high production.")
if avg_sucrose_2 > 14:
    insights_2.append("High sucrose content indicates quality cane, boosting output.")
results['highest_production_2024'] = {
    'factory': top_factory,
    'total_production': round(top_prod_value, 2),
    'insights': insights_2,
    'details': details_2.to_dict(orient='records')
}
markdown_output += f"""
## 2. Highest Production Factory in 2024
- **Factory**: {top_factory}
- **Total Production**: {round(top_prod_value, 2)} tonnes
- **Insights**:
{''.join([f'  - {i}\n' for i in insights_2])}
- **Details**:
{details_2.to_markdown(index=False)}
\n"""

# Insight 3: Sucrose content correlation with production in 2025
corr_sucrose_prod = float(df_2025[['sucrose content', 'Production Quantity (tonnes)']].corr().iloc[0, 1])  # Convert to float
insights_3 = []
if corr_sucrose_prod > 0.5:
    insights_3.append("Strong positive correlation suggests higher sucrose content drives production.")
elif corr_sucrose_prod < 0:
    insights_3.append("Negative correlation may indicate other factors dominate production.")
results['sucrose_production_correlation_2025'] = {
    'correlation': round(corr_sucrose_prod, 2),
    'insights': insights_3
}
markdown_output += f"""
## 3. Sucrose Content Correlation with Production in 2025
- **Correlation**: {round(corr_sucrose_prod, 2)}
- **Insights**:
{''.join([f'  - {i}\n' for i in insights_3])}
\n"""

# Insight 4: Highest yield variety in 2024
avg_yield_by_variety = df_2024.groupby('variety')['Crop Yield (tonnes/ha)'].mean().sort_values(ascending=False)
top_variety = avg_yield_by_variety.index[0]
top_yield_value = float(avg_yield_by_variety.iloc[0])  # Convert to float
details_4 = df_2024[df_2024['variety'] == top_variety][['factory', 'region', 'sucrose content', 'Crop Yield (tonnes/ha)']]
avg_sucrose_4 = float(details_4['sucrose content'].mean(skipna=True))
insights_4 = []
if top_yield_value > 80:
    insights_4.append("High yield suggests this variety is well-suited to local conditions.")
if avg_sucrose_4 > 14:
    insights_4.append("High sucrose content enhances variety performance.")
results['highest_yield_variety_2024'] = {
    'variety': top_variety,
    'average_yield': round(top_yield_value, 2),
    'insights': insights_4,
    'details': details_4.to_dict(orient='records')
}
markdown_output += f"""
## 4. Highest Yield Variety in 2024
- **Variety**: {top_variety}
- **Average Yield**: {round(top_yield_value, 2)} tonnes/ha
- **Insights**:
{''.join([f'  - {i}\n' for i in insights_4])}
- **Details**:
{details_4.to_markdown(index=False)}
\n"""

# Insight 5: Highest production value region in 2025
total_value_by_region = df_2025.groupby('region')['Production Value (KSh million)'].sum().sort_values(ascending=False)
top_value_region = total_value_by_region.index[0]
top_value = float(total_value_by_region.iloc[0])  # Convert to float
details_5 = df_2025[df_2025['region'] == top_value_region][['factory', 'Production Quantity (tonnes)', 'Crop Yield (tonnes/ha)']]
avg_yield_5 = float(details_5['Crop Yield (tonnes/ha)'].mean(skipna=True))
insights_5 = []
if avg_yield_5 > 80:
    insights_5.append("High crop yield supports high economic output.")
results['highest_production_value_2025'] = {
    'region': top_value_region,
    'total_value': round(top_value, 2),
    'insights': insights_5,
    'details': details_5.to_dict(orient='records')
}
markdown_output += f"""
## 5. Highest Production Value Region in 2025
- **Region**: {top_value_region}
- **Total Value**: {round(top_value, 2)} KSh million
- **Insights**:
{''.join([f'  - {i}\n' for i in insights_5])}
- **Details**:
{details_5.to_markdown(index=False)}
\n"""

# Insight 6: Lowest national production weeks in 2025
national_2025 = df_2025[df_2025['region'] == 'National']
lowest_prod_weeks = national_2025.nsmallest(3, 'Production Quantity (tonnes)')[['week', 'Production Quantity (tonnes)', 'Crop Yield (tonnes/ha)']]
avg_yield_6 = float(lowest_prod_weeks['Crop Yield (tonnes/ha)'].mean(skipna=True))
insights_6 = []
if avg_yield_6 < 50:
    insights_6.append("Low yields in these weeks may reflect seasonal or supply issues.")
results['lowest_production_weeks_2025'] = {
    'weeks': lowest_prod_weeks.to_dict(orient='records'),
    'insights': insights_6
}
markdown_output += f"""
## 6. Lowest National Production Weeks in 2025
- **Weeks**:
{lowest_prod_weeks.to_markdown(index=False)}
- **Insights**:
{''.join([f'  - {i}\n' for i in insights_6])}
\n"""

# Insight 7: Highest production value per hectare in 2024
df_2024['value_per_hectare'] = df_2024['Production Value (KSh million)'] / df_2024['Crop Area (hectares)']
top_value_per_hectare = df_2024.groupby('factory')['value_per_hectare'].mean().sort_values(ascending=False)
top_factory_7 = top_value_per_hectare.index[0]
top_value_per_hectare_value = float(top_value_per_hectare.iloc[0])  # Convert to float
details_7 = df_2024[df_2024['factory'] == top_factory_7][['region', 'Crop Yield (tonnes/ha)', 'sucrose content']]
avg_yield_7 = float(details_7['Crop Yield (tonnes/ha)'].mean(skipna=True))
insights_7 = []
if avg_yield_7 > 80:
    insights_7.append("High yield contributes to efficient land use.")
results['highest_value_per_hectare_2024'] = {
    'factory': top_factory_7,
    'average_value_per_hectare': round(top_value_per_hectare_value, 2),
    'insights': insights_7,
    'details': details_7.to_dict(orient='records')
}
markdown_output += f"""
## 7. Highest Production Value per Hectare in 2024
- **Factory**: {top_factory_7}
- **Average Value per Hectare**: {round(top_value_per_hectare_value, 2)} KSh million/ha
- **Insights**:
{''.join([f'  - {i}\n' for i in insights_7])}
- **Details**:
{details_7.to_markdown(index=False)}
\n"""

# Combine outputs
output = {
    'json': convert_to_native(results),
    'markdown': markdown_output
}

# Return JSON string
print(json.dumps(output, indent=2))
````
